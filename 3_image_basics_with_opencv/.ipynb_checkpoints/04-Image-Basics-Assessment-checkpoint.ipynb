#### TASK: Open the *dog_backpack.jpg* image from the DATA folder and display it in the notebook. Make sure to correct for the RGB order.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../DATA/dog_backpack.jpg")
type(img)
```

```python
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

#### TASK: Flip the image upside down and display it in the notebook.

```python
img_flip = img.copy()
img_flip = cv2.flip(img_flip, 0)
plt.imshow(img_flip)
```

#### TASK: Draw an empty RED rectangle around the dogs face and display the image in the notebook.

```python
img_rectangle = img.copy()
cv2.rectangle(img_rectangle, pt1=(250,400), pt2=(600,700), color=(0,0,255), thickness = 5)
plt.imshow(img_rectangle)
```

#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
img_triangle = img.copy()
verticles = np.array([ [400,400],[250,700],[600,700] ], dtype=np.int32)
pts = verticles.reshape((-1,1,2))
cv2.polylines(img_triangle, [pts], isClosed=True, color=(0,0,255), thickness=5)
plt.imshow(img_triangle)
# cv2.drawContours(img_triangle, pt1=(400,400), pt2=(250,700), pt3=(600,700), color=(255,0,0), thickness=5)
```

### BONUS TASK. Can you figure our how to fill in this triangle? It requires a different function that we didn't show in the lecture! See if you can use google search to find it.

[CLICK ME FOR A DIRECT LINK TO THE HINT](https://docs.opencv.org/3.0-beta/modules/imgproc/doc/drawing_functions.html#fillpoly)

```python
img_fill_triangle = img.copy()
verticles = np.array([ [480,400],[650,700],[250,700] ])
pts = verticles.reshape(-1,1,2)
cv2.fillPoly(img_fill_triangle,[pts],color=(0,0,255))
plt.imshow(img_fill_triangle)
```

#### TASK: (NOTE: YOU WILL NEED TO RUN THIS AS A SCRIPT). Create a script that opens the picture and allows you to draw empty red circles whever you click the RIGHT MOUSE BUTTON DOWN.

```python
img_draw_circle = img.copy()
img_draw_circle = cv2.cvtColor(img_draw_circle, cv2.COLOR_BGR2RGB)
# zmienne
# prawda, gdy przycisk wciśnięty, fałsz gdy przycisk puszczony
drawing = False
ix = -1
iy = -1

# rysowanie
def draw_circle(event, x, y, flags, params):
    global ix, iy, drawing
    
    if event == cv2.EVENT_RBUTTONDOWN:
        drawing = True
        ix, iy = x, y
        cv2.circle(img_draw_circle, (ix,iy), 50, (0,0,255),thickness=5)

cv2.namedWindow(winname="obrazek")
cv2.setMouseCallback("obrazek", draw_circle)


################################
## Pokazanie obrazka w OpenCV ##
################################

while True:
    
    cv2.imshow("obrazek",img_draw_circle)
    
    if cv2.waitKey(1) & 0xFF == 27:
        break

cv2.destroyAllWindows()

```
